```python
import nltk
from nltk.corpus import stopwords
import pandas as pd
import string
```

```python
df=pd.read_csv("training.txt",sep='\t',names=['sentiment','text'])
```

```python
df.head()
```

```python
word_count={}
likelihood={}
pos_count=0
df['text'][0].lower().split(" ")
#list of stop words
stopwordslist=list(stopwords.words('english'))+list(string.punctuation)
```

```python
for i in range(0,len(df)):
    if df['sentiment'][i]==1:
        pos_count+=1
    for c in df['text'][i].lower().split(" "):
        # c should not be in stopwords and should be only a string of letter not alphanumberic or numbers or symbols
        if c not in stopwordslist and c.isalpha():
            if c not in word_count.keys() :
                word_count[c]=[0,0]
                likelihood[c]=0
            if df['sentiment'][i]==1:
                word_count[c][1]+=1
            else:
                word_count[c][0]+=1
len(word_count.keys())
```

```python
word_count
```

```python
#likelihood calculation
def calculate_sentiment(s):
    tokens=s.lower().split(" ")
    posterior_prob=1
    posterior_prob_neg=1
    for c in tokens:
        if c in word_count.keys():
            
            if word_count[c][1] != 0:
                posterior_prob*=(word_count[c][1]*1.0/(word_count[c][0]+word_count[c][1]))
            
            if word_count[c][0] != 0:
                posterior_prob_neg*=(word_count[c][0]*1.0/(word_count[c][0]+word_count[c][1]))
    
    print(str("likelihood product for tokens for pos and neg : ")+str(posterior_prob)+" "+str(posterior_prob_neg))
    #positive sentiment value
    sentiment=posterior_prob*pos_count*1.0/len(df)
    #negative sentiment value
    sentiment_neg=posterior_prob_neg*(1-(pos_count*1.0/len(df)))
    
    print(str(sentiment)+" "+str(sentiment_neg))
    if(sentiment>sentiment_neg):
        return True
    else:
        return False
        
```

```python
s="this is good"
if(calculate_sentiment(s)):
    print("The sentence gives a positive sentiment")
else:
    print("The sentence gives a negative sentiment")
word_count['worst']
```
